```python
plt.figure(figsize= (14,8))
plt.subplot(1,2,1)
sns.histplot(df['Price'],bins = 20,kde = True)
plt.title('Distribution of price')
plt.subplot(1, 2, 2)
sns.scatterplot(x='Area', y='Price', data=df)
plt.title('Scatter plot of Area vs. Price')
plt.show()
```
